```python
#Combine 2 bar plots in one

import numpy as np
```

```python
Bikes = ['Hero', 'Honda', 'Yamaha', 'Suzuki']
speed = [50, 70, 40, 80] #kmph
average = [45, 40, 35, 35] #per/ltr
```

```python
# Create two separate barplots
plt.bar(Bikes,speed,width=.3,label='Bike Speed', color='blue', alpha=0.7)
plt.bar(Bikes,average,width=.3,label='Bike Average', color='orange', alpha=0.7)

plt.xlabel('Bikes')
plt.ylabel('Speed & Average')
plt.title('Combined Barplots')
plt.legend()
plt.show()
```

```python
# Sample data (time periods and corresponding values)
time_periods = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
values = [100, 120, 110, 130, 140, 150]

# Create a line plot
plt.plot(time_periods, values, marker='o', linestyle='-')

# Adding labels and a title
plt.xlabel('Time Period')
plt.ylabel('Value')
plt.title('Trend Analysis Over Time')

# Display the plot
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

```python
# Sample data for 2022 and 2023
years = ['2022', '2023']
values_2022 = [100, 120, 110, 130, 140, 150]
values_2023 = [110, 130, 125, 145, 155, 165]

# Create a line plot for 2022
plt.plot(values_2022, label='2022', marker='o', linestyle='-')

# Create a line plot for 2023
plt.plot(values_2023, label='2023', marker='o', linestyle='-')

# Adding labels and a title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trend Analysis for 2022 and 2023')

# Display a legend to differentiate the years
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

#5. Horizontal bar plot: Category vs count plot

```python
items = ['Tea','Samosa','Vadapav','Pohe','Misal']
price = [10,20,25,30,50]
```

```python
plt.barh(items,price)
# to rotate x ticks
plt.ylabel('Items')
plt.xlabel('Price in Rs.')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

#6.Box plot: It is used to check statistical information + it detects outliers if any

```python
age = np.random.randint(2,80,50)
age
```

```python
plt.boxplot(age,vert=False); # ;= show()
```

```python
#lets add some outliers in age
new_age = np.append(age,[-10,-20,150,175,210])
new_age
```

```python
plt.boxplot(new_age,vert=False);
```

```python
#(new_age>1) or (new_age<=110) Wrong approach we ar using logical or

# instead use bitwise &

result = (new_age>1) & (new_age<=110)
result
```

```python
new_age
```

```python
new_age[result]
```

```python
len(new_age)
```

#Sub Plotting

```python
plt.subplot(121) # 1 row, 2 column, 1st subplot
plt.subplot(122) # 1 row, 2 column, 2nd subplot
```

```python
plt.subplot(221) # 2 row, 2 column, 1st subplot
plt.subplot(222) # 2 row, 2 column, 2nd subplot
plt.subplot(223) # 2 row, 2 column, 3rd subplot
plt.subplot(224) # 2 row, 2 column, 4th subplot
```

```python
plt.figure(figsize=(12,8))

plt.subplot(221)
plt.bar(items,price,color='red')
plt.title('Item vs Price plot 1')

plt.subplot(222)
plt.boxplot(age)
plt.title('Boxplot Age')

plt.subplot(224)
plt.barh(items,price,color='g')
plt.title('Item vs Price plot 2')

plt.subplot(223)
plt.hist(new_age,color='y')
plt.title('Histogram newAge')

plt.suptitle('Combining 4 plots')
plt.tight_layout()
#save fig to left side -  files section
plt.savefig('Combine 4 subplots.jpg',dpi=300)
plt.show()
```
