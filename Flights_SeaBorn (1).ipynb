```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('flights')
df.head()
```

#Lets do some SEABORN Plotting over given data

##1.Scatter Plot

```python
sns.scatterplot(x='year',y='passengers',data=df)
```

#2.Hist Plot

```python
df2 = df['passengers']
sns.histplot(data=df2,bins=5,kde=False)
```

#3.Histogram Plot

```python
df.hist(figsize=(8,4))
```

#4.Lm plot

```python
sns.lmplot(x='year',y='passengers', data= df)
```

#5.Joint plot : Kind = Scatter

```python
sns.jointplot(data=df,x='year',y='passengers',kind='scatter')
```

#5.Joint plot kind = Hex

```python
sns.jointplot(data=df,x='year',y='passengers',kind='hex')
```

#5.Joint plot kind = resid

```python
sns.jointplot(data=df,x='year',y='passengers',kind='resid')
```

#5.Joint plot kind = kde

```python
sns.jointplot(data=df,x='year',y='passengers',kind='kde')
```

#5.Joint plot kind = reg

```python
sns.jointplot(data=df,x='year',y='passengers',kind='reg')
```

#6.Kde Plot (Kernel density estimation)

```python
df1 = df['passengers']
sns.kdeplot(data=df1)
```

#7. Pair Plot - works only on numerical data

```python
sns.pairplot(data=df)
```

#8.Box Plot

```python
sns.boxplot(x='year',y='passengers',data=df)
```

#9.Count plot

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df,x='passengers',hue='year')
```

#10.Bar Plot

```python
sns.barplot(x='year',y='passengers',data=df)
```

#11.Strip Plot

```python
sns.stripplot(x='year',y='passengers',data=df)
```

#12.Violin Plot

```python
sns.violinplot(x='year',y='passengers',data=df,palette='Reds')
```

```python
df.corr().style.background_gradient('Greens')
```

#13.heatmap

```python
sns.heatmap(df.corr(),annot=True)
```

#14. Line Plot

```python
sns.lineplot(x='year',y='passengers',data=df)
```
